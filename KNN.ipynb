```python
import pandas as pd
import tkinter.filedialog
from tkinter import *
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

gui = Tk()

gui.title('KNN Classifier GUI')

gui.geometry('600x300')

def data():
    global df
    global check
    filename = askopenfilename(initialdir='C:\\Users\\Desktop\\GUI',title = "Select file") 
    check = StringVar()
    check = "Dataset is uploaded"
    file = open(filename, 'r')
    t_upload.insert(END, check)
    df = pd.read_csv(filename)
def train_data():
    
    testSize=float(splitEntry.get())
    
    try:
        testSize=float(splitEntry.get())

    except ValueError:
        print('Test Size value is not a float')
    
    X = df.iloc[:, :-1]
    y = df.iloc[:, -1]

    global X_train, X_test, y_train, y_test 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=testSize, random_state=42)
    
    global model
    model = KNeighborsClassifier(n_neighbors=7)
    model= model.fit(X_train, y_train)
    
def test_data():
    if not len(t_train.get("1.0", 'end-1c')):
        error_message = "Train the Model First!"
        t_train.delete("1.0", "end")
        t_train.insert(END, error_message)
        return
    global pred
    pred = model.predict(X_test)
        
def res():
    train_data()
    test_data()
    pop=Toplevel()
    pop.geometry("200x300")
    pop.title("Results")
    lbl=Label(pop,text="Accuracy: "+ str(metrics.accuracy_score(y_test,pred)))
    lbl.pack()
    lbl1=Label(pop,text="Precision: "+ str(metrics.precision_score(y_test, pred)))
    lbl1.pack()
    lbl2=Label(pop,text="Recall: "+ str(metrics.recall_score(y_test, pred)))
    lbl2.pack()
    lbl6=Label(pop,text="F1 score: "+ str(metrics.f1_score(y_test,pred)))
    lbl6.pack()
    lbl7=Label(pop,text="Confusion_matrix "+ str(metrics.confusion_matrix(y_test,pred)))
    lbl7.pack()
    pop.mainloop()


LabelFrame1=LabelFrame(gui,text="Loading data")
LabelFrame1.pack(fill="both",ipadx=5,ipady=5)
l1=Label(LabelFrame1, text='Select Data File')
l1.grid(row=0, column=0)
e1=Entry(LabelFrame1,text='')
e1.grid(row=0, column=1)

Button(LabelFrame1,text='open', command=data).grid(row=0, column=2)


paramFrame=LabelFrame(gui,text="Parameters")
paramFrame.pack(fill="both")


Button(paramFrame, text='Train', command=train_data).grid(row=3,column=2)
Button(paramFrame, text='Test',command=test_data).grid(row=3,column=3)

splitLabel=Label(paramFrame, text='Test size:')
splitLabel.grid(row=3,column=0,padx=10,pady=10)

splitEntry=Entry(paramFrame)
splitEntry.grid(row=3, column=1,padx=10,pady=10)

Button(gui, text='Result', command=res).place(x=230,y=230)
Button(gui, text='Close', command=gui.destroy).place(x=300,y=230)


t_upload = Text(gui, width=24, height=1)
t_train = Text(gui, width=20, height=1)

t_upload.place(x=280, y=18)
t_train.place(x=380, y=80)

gui.mainloop()
```
